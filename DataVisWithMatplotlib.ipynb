# Shows how to plot bar charts in matplotlib and label them

```python
import numpy as np
import matplotlib.pyplot as plt
categories = 3
N = 5

cat1_mean = (20, 30, 27, 21, 44)
cat2_mean = (25, 28, 24, 21, 42)
cat3_mean = (19, 30, 33, 20, 41)

cat1_std = (1, 2, 4, 3, 1)
cat2_std = (2, 1, 3, 1, 4)
cat3_std = (2, 1, 3, 3, 1)

x_locs = np.arange(N)
width_of_bars = 0.3

fig, ax = plt.subplots()

rects_cat1 = ax.bar(x_locs, cat1_mean, width_of_bars, color='r', yerr=cat1_std)
rects_cat2 = ax.bar(x_locs + width_of_bars, cat2_mean, width_of_bars, color='y', yerr=cat2_std)
rects_cat3 = ax.bar(x_locs + width_of_bars * 2, cat3_mean, width_of_bars, color='b', yerr=cat3_std)

ax.set_ylabel('Value')
ax.set_xticks(x_locs + width_of_bars )
ax.set_xticklabels(('T1', 'T2', 'T3', 'T4', 'T5'))
ax.legend((rects_cat1[0], rects_cat2[0], rects_cat3[0]), ('Category 1', 'Category 2', 'Category 3'))



def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height, #x and y coordinate
                '%d' % int(height),      #text to be printed
                ha='center', va='bottom') #alignment of the text

autolabel(rects_cat1)
autolabel(rects_cat2)
autolabel(rects_cat3)

plt.show()


```

# Shows how to plot scatter plots

```python
N1 = 300
x1 = np.random.rand(N1)
y1 = (0.2*x1*x1 + 0.1*x1 + 0.4) + 0.01*np.random.randn(N1)

N2 = 300
x2 = np.random.rand(N2)
y2 = 0.1*x2+0.2 + 0.02 *np.random.randn(N2)

fig, ax = plt.subplots()
ax.scatter(x1,y1, c='r', s=10, marker='^')
sorted_x1 = np.sort(x1)
sorted_y1 = (0.2*sorted_x1*sorted_x1 + 0.1*sorted_x1 + 0.4)
ax.plot(sorted_x1, sorted_y1)

ax.scatter(x2,y2, c='g', s=1, marker='o')
sorted_x2 = np.sort(x2)
sorted_y2 = 0.1*sorted_x2+0.2 
ax.plot(sorted_x2, sorted_y2)


import matplotlib.patches as mpatches
red_patch = mpatches.Patch(color='r', label='Polynomial')
green_patch = mpatches.Patch(color='g', label='Linear')
plt.legend(handles=[red_patch, green_patch])

plt.show()
```

```python
N1 = 300
x1 = np.random.rand(N1)
y1 = (0.2*x1*x1 + 0.1*x1 + 0.4) + 0.01*np.random.randn(N1)

N2 = 300
x2 = np.random.rand(N2)
y2 = 0.1*x2+0.2 + 0.02 *np.random.randn(N2)

fig, ax = plt.subplots(1,2)
ax[0].scatter(x1,y1, c='r', s=10, marker='^')
sorted_x1 = np.sort(x1)
sorted_y1 = (0.2*sorted_x1*sorted_x1 + 0.1*sorted_x1 + 0.4)
ax[0].set_title('Polynomial')
ax[0].plot(sorted_x1, sorted_y1)


ax[1].scatter(x2,y2, c='g', s=1, marker='o')
sorted_x2 = np.sort(x2)
sorted_y2 = 0.1*sorted_x2+0.2 
ax[1].plot(sorted_x2, sorted_y2)
ax[1].set_title('Linear')
fig.tight_layout()
plt.show()
```

# Grid Layout and pie chart demo

```python
#Create grid layout
from matplotlib.gridspec import GridSpec
grid_layout = GridSpec(3, 3)

# Create pie chart
labels = ['Cat1', 'Cat2', 'Cat3', 'Cat4', 'Cat5']
area_labels = [0.1, 0.2, 0.35, 0.3, 0.05] #must sum to 1 or 100

plt.subplot(grid_layout[0, 0], aspect=1)
plt.pie(area_labels, labels=labels, shadow=True)

#Another py chart in same row but spanning 2 columns
explode = (0, 0.2, 0, 0, 1)
plt.subplot(grid_layout[0, 1:3])
plt.pie(area_labels, labels=labels, shadow=True, explode=explode, autopct='%d')

#to avoid stretching across columns use aspect = 1
plt.subplot(grid_layout[2, 0:2], aspect = 1)
plt.pie(area_labels, labels=labels, shadow=True, explode=explode, autopct='%d')


#Draw another diagram in next row - scatter
N1 = 300
x1 = np.random.rand(N1)
y1 = (0.2*x1*x1 + 0.1*x1 + 0.4) + 0.01*np.random.randn(N1)
sorted_x1 = np.sort(x1)
sorted_y1 = (0.2*sorted_x1*sorted_x1 + 0.1*sorted_x1 + 0.4)

plt.subplot(grid_layout[1, 0:2])
plt.scatter(x1,y1, c='r', s=10, marker='^')
plt.plot(sorted_x1, sorted_y1)
plt.tight_layout()



plt.subplot(grid_layout[1:3, 2])
y = np.asarray([1,3,5,7,9])
x = np.asarray([3,3,3,3,3])
colors = np.asarray([0.5, 0.2, 0.1, 0.4, 0.5])
circle_area = np.asarray([1, 2, 16, 4, 2])
scale_factor = 100
plt.scatter(x, y, c=colors, s=circle_area*scale_factor, marker='o', alpha=0.5)

plt.show()
```

```python


import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
import matplotlib.dates
from matplotlib.dates import MONTHLY, DateFormatter, rrulewrapper, RRuleLocator
 
from pylab import *
 
def create_date(month,year):
    date = dt.datetime(int(year), int(month), 1)
    mdate = matplotlib.dates.date2num(date)
    return mdate
 
# Data
pos = arange(0,100,5)
 
ylabels = []
ylabels.append('Task1')
ylabels.append('Task2')
ylabels.append('Task3')
ylabels.append('Task4')
ylabels.append('Task5')

 
effort = []
effort.append([0.3, 0.3])
effort.append([0.4, 0.6])
effort.append([0.2, 0.7])
effort.append([0.1, 0.5])
effort.append([0.5, 1])

customDates = []
customDates.append([create_date(5,2014),create_date(6,2014)])
customDates.append([create_date(6,2014),create_date(8,2014),create_date(8,2014)])
customDates.append([create_date(7,2014),create_date(9,2014),create_date(9,2014)])
customDates.append([create_date(10,2014),create_date(3,2015),create_date(3,2015)])
customDates.append([create_date(2,2015),create_date(6,2015),create_date(6,2015)])

task_dates = {}
for i,task in enumerate(ylabels):
    task_dates[task] = customDates[i]

#Initialise plot
fig = plt.figure()
ax = fig.add_subplot(111)
 
# Plot the data
start_date,end_date = task_dates[ylabels[0]]
width_of_outer_bar = 10
width_of_inner_bar = width_of_outer_bar/2.5

y_tics = []

ax.barh(0, end_date - start_date, 
        left=start_date, 
        height=width_of_outer_bar, 
        align='center', color='blue', alpha = 0.75)
y_tics.append(0)
ax.barh(-2.5, (end_date - start_date) * effort[0][0], 
        left=start_date, 
        height=width_of_inner_bar, 
        align='center', color='red', alpha = 0.75, label = "Resource1 Effort")
ax.barh(2.5, (end_date - start_date) * effort[0][1], 
        left=start_date, 
        height=width_of_inner_bar, 
        align='center', color='yellow', alpha = 0.75, label = "Resource2 Effort")



gap_between_bars = 10
for i in range(1,len(ylabels)):
    start_date, mid_date, end_date = task_dates[ylabels[i]]
    piEffort, studentEffort = effort[i]
    y_tics.append((i*5) + i*gap_between_bars) # save tic positions for displaying y labels
    ax.barh((i*5) + i * gap_between_bars, mid_date - start_date, 
            left=start_date, 
            height=width_of_outer_bar, 
            align='center', color='blue', alpha = 0.75)
    ax.barh((i*5) - 2.5 + i*gap_between_bars, (mid_date - start_date) * piEffort, 
            left=start_date, 
            height=width_of_inner_bar, 
            align='center', color='red', alpha = 0.75)
    ax.barh((i*5) + 2.5 + i*gap_between_bars, (mid_date - start_date)*studentEffort, 
            left=start_date, 
            height=width_of_inner_bar, 
            align='center', color='yellow', alpha = 0.75)

#Display dotted grid
ax.grid(color = 'g', linestyle = ':')
ax.axis('tight')

# Format the y-axis
ax.set_yticks(y_tics)
ax.set_yticklabels(ylabels)
ax.set_ylim(ymin = y_tics[0]-5, ymax =  y_tics[-1]+5)
 
# Format the x-axis
ax.xaxis_date() #Tell matplotlib that these are dates...
 
rule = rrulewrapper(MONTHLY, interval=1)
loc = RRuleLocator(rule)
formatter = DateFormatter("%b '%y")
 
ax.xaxis.set_major_locator(loc)
ax.xaxis.set_major_formatter(formatter)
labelsx = ax.get_xticklabels()
plt.setp(labelsx,  rotation=30, fontsize=10)
 
# Format the legend
 
font = font_manager.FontProperties(size='small')
ax.legend(loc=1,prop=font)
 
# Finish up
ax.invert_yaxis()
fig.autofmt_xdate()
plt.show()
```

# Histogram example along with plots within plots

```python
# create some data to use for the plot
dt = 0.001
t = np.arange(0.0, 10.0, dt)
#signal - assume it decays exponentially over time
r = np.exp(-t[:1000] / 0.05)  
#create a gaussian noise
x = np.random.randn(len(t))

#convolve noise with signal
s = np.convolve(x, r)[:len(x)] * dt  # colored noise

# the main axes is subplot(111) by default
plt.plot(t, s)
plt.axis([0, 1, 1 * np.min(s), 3 * np.max(s)])
plt.xlabel('time (s)')
plt.ylabel('current (nA)')
plt.title('Gaussian colored noise')

# this is an inset axes over the main axes
a = plt.axes([0.2, 0.6, .2, .2], facecolor='k')
plt.plot(t[:len(r)], r)
plt.title('Impulse response')
plt.yticks([])

# this is another inset axes over the main axes
a = plt.axes([.65, .6, .2, .2], facecolor='k')
n, bins, patches = plt.hist(s, 400, density=True)
plt.title('Probability')
plt.xticks([-0.1,0,0.1])
plt.yticks([])

plt.show()
```
